```python
# Author: Abdessamad Bensaad
# Link: https://github.com/B3ns44d/Python_Sorting_Algorithms

def merge_sort1(unsorted_list):
    if len(unsorted_list) <= 1:
        return unsorted_list
# Find the middle point and divide the unsorted list
    middle = len(unsorted_list) // 2
    left_list = unsorted_list[:middle]
    right_list = unsorted_list[middle:]

    left_list = merge_sort1(left_list)
    right_list = merge_sort1(right_list)
    return list(merge(left_list, right_list))

# Merge the sorted halves

def merge(left_half,right_half):

    res = []
    while len(left_half) != 0 and len(right_half) != 0:
        if left_half[0] < right_half[0]:
            res.append(left_half[0])
            del left_half[0]
        else:
            res.append(right_half[0])
            del right_half[0]
    if len(left_half) == 0:
        res = res + right_half
    else:
        res = res + left_half
    return res
```

```python
# Author: Abdessamad Bensaad
# Link: https://github.com/B3ns44d/Python_Sorting_Algorithms

def insertion_sort1(InputList):
    for i in range(1, len(InputList)):
        j = i - 1
        nxt_element = InputList[i]
        # Compare the current element with next one

        while (InputList[j] > nxt_element) and (j >= 0):
            InputList[j + 1] = InputList[j]
            j = j - 1
        InputList[j + 1] = nxt_element
```

```python
# Author: Tarcisio Marinho
# Link: https://github.com/tarcisio-marinho/sorting-algorithms/tree/master

def merge_sort2(obj):
    if len(obj) > 1:
        mid = len(obj)//2
        left = obj[:mid]
        right = obj[mid:]

        merge_sort2(left)
        merge_sort2(right)

        i = 0
        j = 0
        k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                obj[k] = left[i]
                i += 1
            else:
                obj[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            obj[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            obj[k] = right[j]
            j += 1
            k += 1
```

```python
import time
import matplotlib.pyplot as plt
import random
```

```python
lst = []
mtime_lst1 = []
mtime_lst2 = []
for i in range(1000):
    lst.append(i)
    random.shuffle(lst)
    
    
    start = time.perf_counter()
    merge_sort1(lst)
    end = time.perf_counter()

    tot_time = end - start
    mtime_lst1.append(tot_time)
    
    random.shuffle(lst)
    
    start = time.perf_counter()
    merge_sort2(lst)
    end = time.perf_counter()

    tot_time = end - start
    mtime_lst2.append(tot_time)
    
    
```

```python
plt.plot(mtime_lst1, label='Merge 1')
plt.plot(mtime_lst2, label='Merge 2')
plt.xlabel('input size')
plt.ylabel('time')
plt.title('Merge Sort Implementation Comparison')
plt.legend()
```

```python
lst = []
itime_lst1 = []
itime_lst2 = []
for i in range(1000):
    lst.append(i)
    random.shuffle(lst)
    
    
    start = time.perf_counter()
    insertion_sort1(lst)
    end = time.perf_counter()

    tot_time = end - start
    itime_lst1.append(tot_time)
    
    random.shuffle(lst)
    
    start = time.perf_counter()
    insertion_sort2(lst)
    end = time.perf_counter()

    tot_time = end - start
    itime_lst2.append(tot_time)
```

```python
plt.plot(itime_lst1, label='Insert 1')
plt.plot(itime_lst2, label='Insert 2')
plt.xlabel('input size')
plt.ylabel('time')
plt.title('Insert Sort Implementation Comparison')
plt.legend()
```

```python
plt.plot(mtime_lst1, label='Merge')
plt.plot(itime_lst1, label='Insert')
plt.xlabel('input size')
plt.ylabel('time')
plt.title('Insert Sort vs. Merge Sort')
plt.legend()
```
